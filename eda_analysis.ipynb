```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train.drop(['label'], axis = 1), train.label, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 200, C = 0.8, n_jobs = 50)
model.fit(X_train, y_train)
model.score(X_val, y_val)
```

```python
test_pred = model.predict(test)
print(test_pred)
```

```python
# Save test predictions to file
output = pd.DataFrame({'ImageId': list(range(1,len(test_pred)+1)),
                       'Label': test_pred})
output.set_index('ImageId', inplace=True)
output.head()
```

```python
output.to_csv("submission.csv", index=True, header=True)
```
